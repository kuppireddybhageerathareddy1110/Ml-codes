```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Load the hiring dataset
hiring_data = pd.read_csv('hiring.csv')

# Display the first few rows and the columns of the dataset
print(hiring_data.head())
print("Columns in the dataset:", hiring_data.columns.tolist())

# Check unique values in the experience column
unique_experience_values = hiring_data['experience'].unique()
print("Unique values in 'experience' column:", unique_experience_values)

# Update experience mapping to include 'eleven'
experience_mapping = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11
}

# Replace values and handle unknowns
hiring_data['experience'] = hiring_data['experience'].str.lower().replace(experience_mapping)
hiring_data['experience'] = pd.to_numeric(hiring_data['experience'], errors='coerce')  # Convert to numeric

# Handle missing values by dropping rows with NaN values
hiring_data.dropna(inplace=True)

# Display the dataset after preprocessing
print(hiring_data.head())
print("Columns after preprocessing:", hiring_data.columns.tolist())

# Create a binary target variable based on salary
salary_threshold = 60000  # Define a threshold
hiring_data['salary'] = (hiring_data['salary($)'] > salary_threshold).astype(int)  # 1 if above threshold, else 0

# Split the dataset into features (X) and target (y)
X = hiring_data.drop(columns=['salary($)', 'salary'])  # Drop original salary column and the new target
y = hiring_data['salary']  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
base_estimator = DecisionTreeClassifier(max_depth=1)

# AdaBoost Classifier
adaboost = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)
adaboost.fit(X_train, y_train)
y_pred_adaboost = adaboost.predict(X_test)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
print(f"AdaBoost Accuracy: {accuracy_adaboost:.2f}")
print("AdaBoost Classification Report:")
print(classification_report(y_test, y_pred_adaboost))

# Gradient Boosting Classifier
gradient_boosting = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gradient_boosting.fit(X_train, y_train)
y_pred_gradient_boosting = gradient_boosting.predict(X_test)
accuracy_gradient_boosting = accuracy_score(y_test, y_pred_gradient_boosting)
print(f"Gradient Boosting Accuracy: {accuracy_gradient_boosting:.2f}")
print("Gradient Boosting Classification Report:")
print(classification_report(y_test, y_pred_gradient_boosting))

# K-means clustering example
# Generate synthetic dataset for clustering (using experience and test score for demonstration)
X_blob, y_blob = make_blobs(n_samples=300, centers=5, cluster_std=1.0, random_state=42)

# Initialize the KMeans model
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model and predict cluster labels
y_kmeans = kmeans.fit_predict(X_blob)

# Plot the clusters
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Mark centroids
plt.title("K-means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Load the Titanic dataset
titanic_data = pd.read_csv('titanic.csv')

# Display the first few rows of the dataset
print(titanic_data.head())
print("Columns in the dataset:", titanic_data.columns.tolist())

# Check for missing values
print(titanic_data.isnull().sum())

# Fill missing 'Age' values with the median
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

# Drop 'Cabin' column due to high number of missing values (optional)
titanic_data.drop(columns=['Cabin'], inplace=True)

# Fill missing 'Embarked' values with the most common port (S)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

# Encode 'Sex' column: male=1, female=0
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 1, 'female': 0})

# Encode 'Embarked' column: C=0, Q=1, S=2
titanic_data['Embarked'] = titanic_data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Drop unnecessary columns
titanic_data.drop(columns=['Name', 'Ticket', 'PassengerId'], inplace=True)

# Prepare features and target variable
X = titanic_data.drop(columns=['Survived'])  # Features
y = titanic_data['Survived']  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
base_estimator = DecisionTreeClassifier(max_depth=1)

# AdaBoost Classifier
adaboost = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)
adaboost.fit(X_train, y_train)
y_pred_adaboost = adaboost.predict(X_test)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
print(f"AdaBoost Accuracy: {accuracy_adaboost:.2f}")
print("AdaBoost Classification Report:")
print(classification_report(y_test, y_pred_adaboost))

# Gradient Boosting Classifier
gradient_boosting = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gradient_boosting.fit(X_train, y_train)
y_pred_gradient_boosting = gradient_boosting.predict(X_test)
accuracy_gradient_boosting = accuracy_score(y_test, y_pred_gradient_boosting)
print(f"Gradient Boosting Accuracy: {accuracy_gradient_boosting:.2f}")
print("Gradient Boosting Classification Report:")
print(classification_report(y_test, y_pred_gradient_boosting))

# K-means clustering example (for visualization purposes)
# Use a subset of features for clustering (Age, Fare, and Pclass for demonstration)
X_cluster = titanic_data[['Age', 'Fare', 'Pclass']].dropna()  # Drop rows with NaN values
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X_cluster)

# Plot the clusters
plt.scatter(X_cluster['Age'], X_cluster['Fare'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Mark centroids
plt.title("K-means Clustering on Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
