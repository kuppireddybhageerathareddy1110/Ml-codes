```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder

# Load the Titanic dataset
df = pd.read_csv('titanic.csv')

# Data Preprocessing
# Fill missing values in 'Age' and 'Fare' columns with the median
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# Label Encoding
le = LabelEncoder()
df['Pclass'] = le.fit_transform(df['Pclass'].astype(str))  # Convert 'Pclass' to string if it's not
df['Sex'] = le.fit_transform(df['Sex'])

# 'Age' and 'Fare' are continuous, so no need for Label Encoding
# They can be used directly in the model

# Select features and target
X = df[['Pclass', 'Sex', 'Age', 'Fare']]
y = df['Survived']

# Build and train the Decision Tree model
model = tree.DecisionTreeClassifier()
model.fit(X, y)

# Evaluate the model
accuracy = model.score(X, y)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Predictions
# Make sure to use the same format for predictions
print("Prediction for [0,0,51,217]:", model.predict(pd.DataFrame([[0, 0, 51, 217]], columns=X.columns)))
print("Prediction for [2,1,28,18]:", model.predict(pd.DataFrame([[2, 1, 28, 18]], columns=X.columns)))
```
